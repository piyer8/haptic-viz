```python
import glob
import numpy as np
from scipy.io import wavfile
from scipy import interpolate
import matplotlib.pyplot as plt
import os
```

```python
# utility function
def save_waveplot(data, folder, signalname, isToSave):
    fig = plt.figure(1, figsize=(16,3))
    plt.plot(data)
    plt.title(signalname)
    if isToSave:
        plt.savefig(folder + signalname+ '.png')
        plt.close(fig)
    else:
        plt.show() 
```

```python
# Function to process all .wav files
def process_wav_files(folder=".", isToSave=True):
    # Get all .wav files in the current folder
    wav_files = glob.glob(os.path.join(folder, "*.wav"))

    # Ensure output folder exists if saving plots
    output_folder = os.path.join(folder, "waveplots/")
    if isToSave and not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Process each .wav file
    for wav_file in wav_files:
        try:
            # Read the .wav file
            sample_rate, data = wavfile.read(wav_file)
            
            # Extract the filename without extension
            signalname = os.path.splitext(os.path.basename(wav_file))[0]

            # Save or show waveplot
            save_waveplot(data, output_folder if isToSave else "", signalname, isToSave)
            print(f"Processed: {wav_file}")
        
        except Exception as e:
            print(f"Error processing {wav_file}: {e}")

# Run the script
if __name__ == "__main__":
    process_wav_files(isToSave=True)  # Set to False if you want to display plots instead
```
