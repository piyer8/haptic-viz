```python
features = ['mean_amplitude', 'rms', 'pulse_count', 'zero_count',
       'mean_onset_strength', 'spectral_centroid']

mean_amplitude = []
rms = []
pulse_count = []
std_pulse_dist = []
zero_count = []
mean_onset_strength = []
spectral_centroid = []

feature_lists = [mean_amplitude, rms, pulse_count, zero_count, mean_onset_strength, spectral_centroid]


for f in files:
    y, sr = rs.load(f)

    onset_env = rs.onset.onset_strength(y=y, sr=sr)
    pulse = rs.beat.plp(onset_envelope=onset_env, sr=sr)
    beats_plp = np.flatnonzero(rs.util.localmax(pulse))
    z = rs.zero_crossings(y)
    cent = rs.feature.spectral_centroid(y=y, sr=sr)
    
    mean_amplitude.append(y.mean())
    rms.append(rs.feature.rms(y=y).mean())
    pulse_count.append(len(beats_plp))
    zero_count.append(np.nonzero(z)[0].size)    
    mean_onset_strength.append(onset_env.mean())
    spectral_centroid.append(cent.mean())
    

```

```python
for i, f in enumerate(features):
    df_features[f] = feature_lists[i]
    
```
