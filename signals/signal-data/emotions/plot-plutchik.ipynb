```python
from pyplutchik import plutchik
import json
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
```

```python
def plot_plutchik(signal_path):
    with open(signal_path) as f:
        signal = json.load(f)
    signal_id = signal[0]['signal_id']
    keyword_data = signal[0]['emotional_keywords']
    emotion_values = {
        'joy': 0,
        'trust': 0,
        'fear': 0,
        'surprise': 0,
        'sadness': 0,
        'disgust': 0,
        'anger': 0,
        'anticipation': 0
    }
    emotion_count = {
        'joy': 0,
        'trust': 0,
        'fear': 0,
        'surprise': 0,
        'sadness': 0,
        'disgust': 0,
        'anger': 0,
        'anticipation': 0
    }
    total_count = 0
    total_intensity = 0
    for keyword in keyword_data:
        emotion = keyword['emotion']
        intensity = float(keyword['intensity'])
        emotion_values[emotion] += intensity
        emotion_count[emotion] += 1
        total_count += 1
        total_intensity += intensity
    
    for emotion in emotion_values:
        if(emotion_count[emotion] != 0):
            emotion_values[emotion] /= emotion_count[emotion]
    

    ax = plutchik(emotion_values)
    ax.get_figure().savefig('./signal-plots/F'+signal_id+'.png')
    print("Saved plot to ./signal-plots/F"+signal_id+".png")
    plt.close()
```

```python
def main():
    mypath = './emotional-data/'
    files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    for file in files:
        if file.endswith(".json"):
            signal_path = mypath + file
            plot_plutchik(signal_path)

if __name__ == '__main__':
    main()
```
