```python


np.random.seed(42)
y = np.zeros([32,1])
ids = np.random.choice(range(32), size=16, replace=False)
y[ids] = 1

# Assume X and y are your features and labels
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.3, random_state = 42)
```
